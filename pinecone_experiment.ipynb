```python
from pinecone import Pinecone, ServerlessSpec
import os
```

```python
if not os.environ.get("PINECONE_API_KEU"):
  from pinecone_notebooks.colab import Authenticate
  Authenticate()
```

```python
pc = Pinecone(api_key=os.environ["PINECONE_API_KEY"])
```

```python
pc.create_index(
    name="demo-index",
    dimension=512,
    metric="cosine",
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    )
)
```

```python
index = pc.Index("demo-index")
```

```python
index
```

```python
import numpy as np

vectors = [
    ("vec1", np.random.rand(512), {"text": "I am a well educated."}),
    ("vec2", np.random.rand(512), {"text": "I am illetrate."}),
    ("vec3", np.random.rand(512), {"text": "I am a good person."}),
]
```

```python
index.upsert(vectors)
```

```python
print(index)
```

```python
query_vector = np.random.rand(512)

result = index.query(
    vector = query_vector.tolist(),
    top_k = 2,
    include_values = True,
    include_metadata = True
)
print(result)
```

```python
index.delete(ids=["vec2"])
```

```python
pc.delete_index('demo-index')
```
